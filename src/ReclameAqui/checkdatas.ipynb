````python
from vincicompass import s3_manager as s3
s3 = s3.S3Manager('datascience-studies-files')
import pickle
import pandas as pd
````

````python
file_bytes = s3.get_file_bytes('repo_data/ReclameAqui/all_companies_full.pkl')
all_companies_full = pickle.loads(file_bytes)
all_companies_full
````

````python
all_companies_full
````

````python
# Verificar tipo e estrutura dos dados
print(f"Tipo: {type(all_companies_full)}")
print(f"Tamanho: {len(all_companies_full)}")
if isinstance(all_companies_full, pd.DataFrame):
    print(f"\nShape: {all_companies_full.shape}")
    print(f"\nColunas: {all_companies_full.columns.tolist()}")
    print(f"\nPrimeiras linhas:")
    display(all_companies_full.head())
    print(f"\nInfo:")
    all_companies_full.info()
elif isinstance(all_companies_full, list):
    print(f"\nPrimeiro item: {all_companies_full[0]}")
elif isinstance(all_companies_full, dict):
    print(f"\nChaves: {all_companies_full.keys()}")
````

````python
# Verificar campos que podem servir como identificadores únicos e timestamps
print("Campos de data/timestamp:")
date_cols = [col for col in all_companies_full.columns if 'date' in col.lower() or 'created' in col.lower() or 'modified' in col.lower()]
print(date_cols)

print("\n\nValores únicos de 'id':", all_companies_full['id'].nunique())
print("Total de linhas:", len(all_companies_full))
print("IDs únicos?", all_companies_full['id'].nunique() == len(all_companies_full))

print("\n\nExemplos de campos de data:")
for col in ['created', 'modified']:
    if col in all_companies_full.columns:
        print(f"\n{col}:")
        print(f"  Tipo: {all_companies_full[col].dtype}")
        print(f"  Exemplo: {all_companies_full[col].iloc[0]}")
        print(f"  Valores únicos: {all_companies_full[col].nunique()}")
````

````python
# Verificar se tem IDs duplicados e analisar mais detalhes
print("IDs duplicados:")
duplicated_ids = all_companies_full[all_companies_full['id'].duplicated(keep=False)]
print(f"Registros com IDs duplicados: {len(duplicated_ids)}")
if len(duplicated_ids) > 0:
    print("\nExemplo de ID duplicado:")
    example_id = duplicated_ids['id'].iloc[0]
    display(all_companies_full[all_companies_full['id'] == example_id][['id', 'created', 'modified', 'title', 'companyName']])

# Verificar distribuição temporal
print("\n\nDistribuição temporal dos dados:")
print(f"Data mais antiga: {all_companies_full['created'].min()}")
print(f"Data mais recente: {all_companies_full['created'].max()}")
print(f"\nDistribuição por ano:")
all_companies_full['year'] = pd.to_datetime(all_companies_full['created']).dt.year
print(all_companies_full['year'].value_counts().sort_index())
````

## Teste da Lógica de Atualização Incremental

Vamos simular o processo de identificação de registros novos:

````python
# Simular cenário: MongoDB tem dados até 2025-11-01
# Quantos registros novos teríamos?

import pandas as pd

# Simular última data no MongoDB
simulated_last_date = pd.to_datetime('2025-11-01')

# Filtrar dados novos
df_new = all_companies_full[pd.to_datetime(all_companies_full['created']) > simulated_last_date]

print(f"📊 Simulação de Atualização Incremental")
print(f"{'='*60}")
print(f"Última data no MongoDB (simulada): {simulated_last_date}")
print(f"Total de registros na base S3: {len(all_companies_full):,}")
print(f"Registros novos a processar: {len(df_new):,}")
print(f"Percentual: {100*len(df_new)/len(all_companies_full):.2f}%")
print(f"\nEconomia de processamento: {len(all_companies_full) - len(df_new):,} registros")
print(f"{'='*60}")

# Mostrar distribuição dos novos registros
if len(df_new) > 0:
    print(f"\n📅 Distribuição temporal dos novos registros:")
    df_new['date'] = pd.to_datetime(df_new['created']).dt.date
    print(df_new['date'].value_counts().sort_index())
    
    print(f"\n🏢 Top 10 empresas com mais reclamações novas:")
    print(df_new['companyName'].value_counts().head(10))
````

## ✅ Sistema de Atualização Incremental Implementado

### 📋 Resumo da Solução

O sistema foi projetado para atualizar eficientemente a base de 721k+ registros do ReclameAqui no MongoDB:

#### 🔑 Estratégia Principal
- **Campo único**: `id` (primary key)
- **Timestamp**: `created` (data de criação)
- **Método**: Atualização incremental usando última data no MongoDB

#### 🎯 Benefícios
- ✅ Processa apenas dados novos
- ✅ Evita duplicatas (upsert)
- ✅ Performance otimizada (batch de 1000)
- ✅ Economia de 99%+ de processamento em atualizações

#### 📁 Arquivos Criados

1. **`collector.py`**: Busca dados do S3
2. **`AllCompanies.py`**: Modelo MongoDB com atualização incremental
3. **`run_scraper.py`**: Script principal de atualização
4. **`run_update.sh`**: Script shell para execução
5. **`examples.py`**: 8 exemplos de consultas
6. **`ATUALIZACAO_INCREMENTAL.md`**: Documentação completa

#### 🚀 Como Usar

```bash
# 1. Configurar conexão
export CONNECTION_URL="mongodb://..."

# 2. Executar atualização
./run_update.sh

# ou diretamente:
python src/ReclameAqui/run_scraper.py
```

#### 📊 Performance Esperada
- **Primeira execução**: ~30-60 minutos (721k registros)
- **Atualizações diárias**: ~1-2 minutos (apenas registros novos)
- **Base atualizada**: ~30 segundos (apenas verificação)
