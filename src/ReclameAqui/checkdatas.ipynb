## 🔍 Análise de Campos Complexos

Vamos verificar a estrutura dos campos que são dicionários ou listas:

````python
# Identificar colunas com tipos complexos
complex_fields = ['interactions', 'additionalFields', 'additionalInfo', 'address', 
                  'phones', 'files', 'companyIndexes', 'complainMediaInfos', 'raFormsAnswer']

print("📊 Análise de Campos Complexos")
print("=" * 80)

for field in complex_fields:
    if field in all_companies_full.columns:
        # Pegar um exemplo não-nulo
        non_null = all_companies_full[all_companies_full[field].notna()][field]
        
        if len(non_null) > 0:
            example = non_null.iloc[0]
            print(f"\n🔹 Campo: {field}")
            print(f"   Tipo: {type(example)}")
            print(f"   Não-nulos: {all_companies_full[field].notna().sum():,} ({100*all_companies_full[field].notna().sum()/len(all_companies_full):.1f}%)")
            
            if isinstance(example, (list, dict)):
                print(f"   Exemplo (primeiros 500 chars):")
                import json
                example_str = json.dumps(example, indent=2, default=str)
                print(f"   {example_str[:500]}...")
            else:
                print(f"   Exemplo: {str(example)[:200]}")
        else:
            print(f"\n🔹 Campo: {field}")
            print(f"   Todos valores são nulos")
    else:
        print(f"\n🔹 Campo: {field}")
        print(f"   ⚠️ Campo não existe no DataFrame")
````

````python
# Pegar um registro com interactions não-nulo
sample_row = all_companies_full[all_companies_full['interactions'].notna()].iloc[0]

print("📋 Exemplo de Registro Completo")
print("=" * 80)
print(f"ID: {sample_row['id']}")
print(f"Empresa: {sample_row['companyName']}")
print(f"Título: {sample_row['title']}")
print(f"Data: {sample_row['created']}")
print(f"\n" + "=" * 80)
print("🔹 Campo 'interactions' (lista de dicts):")
print(f"   Tipo: {type(sample_row['interactions'])}")
print(f"   Tamanho: {len(sample_row['interactions']) if isinstance(sample_row['interactions'], list) else 'N/A'}")

if isinstance(sample_row['interactions'], list) and len(sample_row['interactions']) > 0:
    import json
    print(f"\n   Primeira interação:")
    print(json.dumps(sample_row['interactions'][0], indent=4, default=str))
    
print(f"\n" + "=" * 80)
print("🔹 Outros campos complexos:")

for field in ['address', 'phones', 'files', 'companyIndexes']:
    try:
        val = sample_row.get(field)
        if val is not None and not (isinstance(val, float) and pd.isna(val)):
            if isinstance(val, list):
                print(f"   {field}: lista com {len(val)} itens")
            elif isinstance(val, dict):
                print(f"   {field}: dict com {len(val)} chaves")
            else:
                print(f"   {field}: {type(val)}")
    except:
        pass
````

### ✅ Ajustes Realizados no Modelo

Com base na análise dos campos complexos, foram feitos os seguintes ajustes:

#### 🔧 Campos Corrigidos:

1. **`interactions`**: Alterado de `IntField()` para `ListField()`
   - Armazena lista de dicts com histórico de interações/respostas
   - Cada interação contém: id, created, deleted, message, type, user, etc.

2. **`additionalInfo`**: Alterado de `DictField()` para `StringField()`
   - Na base real, é uma string (não um dicionário)

3. **`companyIndexes`**: Mantido como `ListField()`
   - Lista de strings (JSON serializado) com métricas da empresa

#### 📊 Estatísticas dos Campos Complexos:

| Campo | Tipo | % Preenchido | Observação |
|-------|------|--------------|------------|
| `interactions` | list[dict] | 93.4% | Histórico de respostas |
| `address` | dict | 93.4% | Geralmente vazio {} |
| `phones` | list | 91.6% | Geralmente vazia [] |
| `files` | list | 91.6% | Anexos da reclamação |
| `companyIndexes` | list[str] | 1.7% | Métricas serializadas |
| `additionalInfo` | str | 9.1% | String livre |

#### 🎯 Tratamento Implementado:

- ✅ Validação de tipo antes de inserir
- ✅ Listas e dicts vazios não são inseridos (economia de espaço)
- ✅ Tentativa de parse JSON para strings
- ✅ Remoção de valores None antes do upsert

````python
# Buscar todos os valores distintos de empresa_origem
empresas_origem = AllCompanies.objects.distinct('empresa_origem')

print(f"📊 Valores distintos de 'empresa_origem':")
print(f"   Total: {len(empresas_origem)}\n")

# Ordenar e exibir
empresas_origem_sorted = sorted([e for e in empresas_origem if e is not None])

print("Lista de empresas:")
print("=" * 80)
for i, empresa in enumerate(empresas_origem_sorted, 1):
    print(f"{i:4}. {empresa}")

# Verificar quantos registros são None/null
total_records = AllCompanies.objects.count()
records_with_origem = AllCompanies.objects(empresa_origem__ne=None).count()
records_without_origem = total_records - records_with_origem

print(f"\n" + "=" * 80)
print(f"Estatísticas:")
print(f"   Total de registros: {total_records:,}")
print(f"   Com empresa_origem: {records_with_origem:,} ({100*records_with_origem/total_records:.1f}%)")
print(f"   Sem empresa_origem (None/null): {records_without_origem:,} ({100*records_without_origem/total_records:.1f}%)")
````
